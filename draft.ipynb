```python
import matplotlib.pyplot as plt
import numpy as np

# 
model1_preds = np.random.rand(100, 2)
model2_preds = np.random.rand(100, 2)
model3_preds = np.random.rand(100, 2)
model4_preds = np.random.rand(100, 2)

# 
fig, ax = plt.subplots(figsize=(10, 10))

# 
ax.scatter(model1_preds[:, 0], model1_preds[:, 1], color='r', marker='o', label='Model 1')
ax.scatter(model2_preds[:, 0], model2_preds[:, 1], color='g', marker='^', label='Model 2')
ax.scatter(model3_preds[:, 0], model3_preds[:, 1], color='b', marker='s', label='Model 3')
ax.scatter(model4_preds[:, 0], model4_preds[:, 1], color='m', marker='D', label='Model 4')

# 
ax.set_title('All Models Predictions')
ax.legend()

# 
plt.show()
```
