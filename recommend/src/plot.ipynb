```python
import matplotlib.pyplot as plt
import json
```

```python
file_paths = ['./goodbooks_2024-06-28_23-18',
              './goodbooks_2024-06-29_18-37'
             ]

labels = ['learnable', 'fixed']
```

```python
# epochloss
all_losses = []

# matplotliblossepoch
plt.figure(figsize=(10, 5))

# 
for i in range(len(file_paths)):
    epochs = []
    losses = []
    with open(file_paths[i] + '/train_log.jsonl', 'r') as file:
        for line in file:
            data = json.loads(line)
            epochs.append(data['epoch'])
            losses.append(data['loss'])
    all_losses.append(losses)

for i, losses in enumerate(all_losses):
    plt.plot(range(1, 61), losses[:60], label=labels[i])

plt.title('Train Loss / Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
